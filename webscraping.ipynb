```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.support.wait import WebDriverWait
```

```python
driver = webdriver.Chrome()
driver.get('https://www.nba.com/stats/players/traditional?Season=2023-24')
```

```python
df2 = pd.DataFrame()
```

```python
table = driver.find_element(By.CSS_SELECTOR, 'table[class="Crom_table__p1iZz"]')
```

```python
while True:
    a = [[i.text for i in j.find_elements(By.CSS_SELECTOR, 'td')] for j in table.find_elements(By.CSS_SELECTOR, 'tr')]
    df2 = pd.concat((df2, pd.DataFrame(a)))
    try:
        next_page = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[data-pos='next']"))).click()
    except:
        break
    print(df2.shape)
```

```python
df2 = df2[1:]
```

```python
df2 = df2.rename(columns={
    0: 'Rank'
    1: 'Player',
    2: 'Team',
    3: 'Age',
    4: 'GP',
    5: 'Wins',
    6: 'Losses',
    7: 'MIN',
    8: 'PTS',
    9: 'FGM',
    10: 'FGA',
    11: 'FG%',
    12: '3PM',
    13: '3PA',
    14: '3P%',
    15: 'FTM',
    16: 'FTA',
    17: 'FT%',
    18: 'OREB',
    19: 'DREB',
    20: 'REB',
    21: 'AST',
    22: 'TOV',
    23: 'STL',
    24: 'BLK',
    25: 'PF',
    26: 'FP',
    27: 'DD2',
    28: 'TD3',
    29: 'PlusMinus'
})
```

```python
df2 = df2.set_index('Rank')
```

```python
df2.iloc[:, 2:6] = df2.iloc[:, 2:6].astype(pd.Int64Dtype())
```

```python
df2.iloc[:, 6:] = df2.iloc[:, 6:].astype(pd.Float64Dtype())
```
