```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.support.wait import WebDriverWait
```

```python
driver = webdriver.Chrome()
driver.get('https://www.nba.com/stats/alltime-leaders')
```

```python
table = driver.find_element(By.CSS_SELECTOR, 'table[class="Crom_table__p1iZz"]')
```

```python
df = pd.DataFrame()
```

```python
while True:
    a = [[i.text for i in j.find_elements(By.CSS_SELECTOR, 'td')] for j in table.find_elements(By.CSS_SELECTOR, 'tr')]
    df = pd.concat((df, pd.DataFrame(a)))
    try:
        next_page = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[data-pos='next']"))).click()
    except:
        break
    print(df.shape)
```

```python
df
```

```python
df = df.rename(columns={
    0: '#',
    1: 'Player',
    2: 'GP',
    3: 'MIN',
    4: 'PTS',
    5: 'FGM',
    6: 'FGA',
    7: 'FG%',
    8: '3PM',
    9: '3PA',
    10: '3P%',
    11: 'FTM',
    12: 'FTA',
    13: 'FT%',
    14: 'OREB',
    15: 'DREB',
    16: 'REB',
    17: 'AST',
    18: 'STL',
    19: 'BLK',
    20: 'TOV',
    21: 'EFG%',
    22: 'TS%'
})
```

```python
df = df[1:]
```

```python
df
```

```python
df.loc[:, 'GP'] = df['GP'].str.replace(',', '')
```

```python
pd.to_numeric(df['#'])
```

```python
df.iloc[:, 0] = df.iloc[:, 0].astype(pd.Int64Dtype())
```

```python
df.columns
```

```python
df = df.replace('-', None)
```

```python
df.iloc[:, 2:7] = df.iloc[:, 2:7].astype(pd.Int64Dtype())
```

```python
df.iloc[:, 7] = df.iloc[:, 7].astype(pd.Float64Dtype())
```

```python
df.iloc[:, 8:10] = df.iloc[:, 8:10].astype(pd.Int64Dtype())
```

```python
df.iloc[:, 10] = df.iloc[:, 10].astype(pd.Float64Dtype())
```

```python
df.iloc[:, 11:13] = df.iloc[:, 11:13].astype(pd.Int64Dtype())
```

```python
df.iloc[:, 13] = df.iloc[:, 13].astype(pd.Float64Dtype())
```

```python
df.iloc[:, 14:] = df.iloc[:, 0].astype(pd.Int64Dtype())
```

```python
df.to_csv('nbadata.csv')
```
